```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
```

```python
n = np.arange(-10, 11, 1)
```

### تابع ضربه گسسته

```python
def D(n):
    return np.piecewise(n, [n<0, 0==n, n>0], [0, 1, 0])
```

```python
plt.stem(n ,D(n) ,use_line_collection=True);
```

### تابع پله گسسته

```python
def U(n):
    return np.piecewise(n, [n<0, 0<=n], [0, 1])
```

```python
plt.stem(n ,U(n) ,use_line_collection=True);
```

## A

### پاسخ ضربه

```python
A_D = 0.5 * (D(n) + D(n-1))
```

```python
plt.stem(n ,A_D ,use_line_collection=True);
```

### پاسخ پله

```python
A_U = 0.5 * (U(n) + U(n-1))
```

```python
plt.stem(n ,A_U ,use_line_collection=True);
```

## B

### پاسخ ضربه

```python
B_D = D(n+1) - D(n)
```

```python
plt.stem(n ,B_D ,use_line_collection=True);
```

### پاسخ پله

```python
B_U = U(n+1) - U(n)
```

```python
plt.stem(n ,B_U ,use_line_collection=True);
```

## C

### پاسخ ضربه

```python
C_D = D(n+1) - 2*D(n) + D(n-1)
```

```python
plt.stem(n ,C_D ,use_line_collection=True);
```

### پاسخ پله

```python
C_U = U(n+1) - 2*U(n) + U(n-1)
```

```python
plt.stem(n ,C_U ,use_line_collection=True);
```

## D

```python
def D_func(X, n):
    m = np.arange(0, 20, 1)
    func = np.cos((np.pi/10) * m) * X(n-m)
    return func.sum()
```

```python
n = np.arange(-10, 30, 1)
```

### پاسخ ضربه

```python
D_D = [D_func(D, i) for i in n]
```

```python
plt.stem(n ,D_D ,use_line_collection=True);
```

### پاسخ پله

```python
D_U = [D_func(U, i) for i in n]
```

```python
plt.stem(n ,D_U ,use_line_collection=True);
```
