```python
import numpy as np
import matplotlib.pyplot as plt
```

## تعریف مقادیر ثابت ها

```python
n = np.arange(0,10,1)
a = 10
k = 0.2
pi = np.pi
T = 5
```

##  تعریف تابع ایکس

```python
def X(n):
    return np.cos(2*k*pi*n)
```

```python
y = X(n)
plt.title("X function")
plt.xlabel("n")
plt.ylabel("X[n]")
plt.stem(n,y,use_line_collection=True);
```

# الف)

## تعریف تابع اول

```python
def A(n):
    return (1/T) * (X(n+1) - X(n))
```

```python
y = A(n)
plt.title("First function")
plt.xlabel("n")
plt.ylabel("y[n]")
plt.stem(n,y,use_line_collection=True);
```

## تعریف تابع دوم

```python
def B(n):
    return (1/(2*T)) * (X(n+1) - X(n-1))
```

```python
y = B(n)
plt.title("Second function")
plt.xlabel("n")
plt.ylabel("y[n]")
plt.stem(n,y,use_line_collection=True);
```

## تعریف تابع سوم

```python
def C(n):
    return (1/(T**2)) * (X(n+1) - 2*X(n) + X(n-1))
```

```python
y = C(n)
plt.title("Third function")
plt.xlabel("n")
plt.ylabel("y[n]")
plt.stem(n,y,use_line_collection=True);
```

## تعریف تابع چهارم

```python
def D(n):
    return 0.1*X(n-2) + 0.2*X(n-1) + 0.4*X(n) + 0.2*X(n+1) + 0.1*X(n+2)
```

```python
y = D(n)
plt.title("Forth function")
plt.xlabel("n")
plt.ylabel("y[n]")
plt.stem(n,y,use_line_collection=True);
```
